```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_2014 = pd.read_csv("../data/neiss2014readable.csv")
data_2019 = pd.read_csv("../data/neiss2019readable.csv")
```

```python
data_2014.head()
```

```python
count2014 = data_2014.groupby("prod_1").size()
count2019 = data_2019.groupby("prod_1").size()
```

```python
count2014 = pd.DataFrame(count2014)
count2019 = pd.DataFrame(count2019)
```

```python
count2014.reset_index(inplace=True)
count2019.reset_index(inplace=True)
```

```python
count2014.columns = ['product','count_2014']
count2019.columns = ['product','count_2019']
```

```python
comb_count = count2014.merge(count2019, on='product')
```

```python
comb_count['increase'] = comb_count['count_2019'] / comb_count['count_2014']
```

```python
comb_count.sort_values(by='increase', ascending=False)
```

```python
comb_count.to_csv('../data/increase.csv', index=False)
```
